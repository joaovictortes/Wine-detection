```python
# Carregando o conjunto de dados
# Link Kaggle: https://www.kaggle.com/dell4010/wine-dataset
import pandas as pd
import numpy as np
import sklearn as sk
arquivo = pd.read_csv('wine_dataset.csv')
```

```python
#Exibir as primeiras linhas do arquivo. PS.: esse tipo de arquivo é um dataframe, tabela com colunas e linhas.
arquivo.head()
```

```python
#Verificando os valores da variável "style" 
arquivo['style'].value_counts()
```

```python
#Convertendo a variável "style" de texto para numérica (Binária). Utilizando uma função:
def transf_style(valor):
    if valor == 'red':
        return 0
    else:
        return 1
#Criando uma nova coluna "style_bin" na dataframe para armazenar a variável style binária definida na função acima
arquivo['style_bin'] = arquivo['style'].map(transf_style)
```

```python
#Outra forma de converter a variável para numérica. Usando a função replace:
arquivo['style'] = arquivo['style'].replace('red', 0)
arquivo['style'] = arquivo['style'].replace('white', 1)

#Neste caso, a coluna style terá os "red" e "white" substituídos por 0 e 1.
```

```python
#Machine Learning: utilizar todas as outras variáveis para detectar se o vinho é branco ou tinto
#Definindo variáveis preditoras (eixo x) e alvo (eixo y). Naturalmente, a variável alvo (dependente) será o "style" e as variáveis independentes serão todas as demais
y = arquivo['style'] #armazenando a coluna de style no eixo y
x = arquivo.drop(columns=['style']) #dropando style da dataframe e armazenando as demais colunas como variáveis do eixo x
```

```python
#Separando as variáveis entre treino e teste. Treino: dados que o modelo usa para aprender. Teste: dados utilizados para avaliar a acurácia do modelo

from sklearn.model_selection import train_test_split #importando a função train_test_split, dentro do pacote model_selection da biblioteca sklearn.

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2)
```

```python
#Executando o machine learning
from sklearn.ensemble import ExtraTreesClassifier #importando o algoritmo/função de ML ExtraTreesClassifier, que está dentro do pacote ensemble, dentro da biblioteca do sklearn

modelo = ExtraTreesClassifier() #definindo a variável "modelo" que recebe a função ExtraTreesClassifier / instanciando a função

modelo.fit(x_treino, y_treino) #treinando os dados, construindo as árvores relacionais

resultado = modelo.score(x_teste, y_teste) #aplicando as árvores relacionais aprendidas no conjunto de dados treino (80% do total). Então exibir o "score" (acurácia)
print(resultado) #O score indica que o algoritmo acertou se o vinho é branco ou tinto em mais de 99% dos dados de teste (20% do total)
```

```python
np.set_printoptions(edgeitems=1000) #Aumentar extensão do output, sem essa linha mostra apenas 3 primeiros e 3 últimos valores
previsao = modelo.predict(x_teste) #Algoritmo prediz os valores de y_teste em relação aos valores de x_teste. Se os vinhos são tinto ou branco no conjunto de dados teste
previsao
```
